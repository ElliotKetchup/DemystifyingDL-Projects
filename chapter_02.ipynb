# 2.1: Implementing "Algorithm 2.1"

```python
import numpy as np
import math

layers = [
    np.array([[-2.74015],[0.93266],[2.93638]]),
    np.array([[-1.09484, 0.74208,  1.48851],
              [-2.76817, -1.37805, 1.3605]]),
    np.array([[-0.53813, -0.65412]])
]

biases = [
    [np.array([[-0.17337], [-2.4558], [-1.677]])],
    [np.array([[-0.53813], [-0.65412]])],
    [np.array([[-0.58768]])]
]

def sigmoid(u):
  return 1/(1+math.exp(-u))

sigmoid_v = np.vectorize(sigmoid)

def ff_ann(input, layers, biases):
  z = np.array([[input]])
  for layer, bias in zip(layers, biases):
    z=(sigmoid_v((layer@z)+bias))

  return z[0][0]

x = 5
ff_ann(x, layers, biases), math.sin(x)
```
